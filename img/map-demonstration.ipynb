```python
# Import 4 different python libraries - pound sign marks this as a comment, unlike the markdown programming in Github
# pandas is a tool we use to import tabular data
# middle two are ??
# osmnx is open street maps 
import pandas as pd
import hvplot.pandas
import holoviews as hv
import osmnx as osm

```

```python
# Search for boundary using OSM
# querying Open Street Maps using the unique identifier, W indicates a "way"
#create a variable name called "ci_boundary"

cu_boundary = osm.geocode_to_gdf(
                        "W46226108",
                        by_osmid=True
)

cu_boundary
```

```python
cu_boundary.plot()
```

```python
# Make interactive map
cu_map = cu_boundary.hvplot(geo = True, tiles = "EsriImagery")
cu_map
```

```python
# Save the map as a file to put on the web (OPTIONAL)
# we have already created the map and now saving it as an html file
hv.save(cu_map, 'cu_map.html')

# Display the map
cu_map
```
